## Checking users 

```python
import re
username_regex =  r'([\w\.]+)'
```

```python
list_of_dataframes = [df_mental_health , df_fitness , df_disorders_1 , df_disorders_2 , df_mental_health_2 , df_fitness_2 , df_fitness_3 , df_disorders_3]
for dataframe in list_of_dataframes:
  dataframe['target'] = dataframe['target'].str.extract(username_regex, expand= True) #drop filas que tengan none después de extracción
  dataframe['source'] = dataframe['source'].str.extract(username_regex, expand=True) #podemos aplicar tranquilamente
```

```python
for change in list_of_dataframes:
  if change['source'].isna().any():
      print('Al menos un elemento en la columna A es None')
  else:
      print('No hay elementos None en la columna A')
```

```python
#df_mental_health_2, df_fitness_2,  df_fitness_2 drop where none after change in target
for change in list_of_dataframes:
  if change['target'].isna().any():
      print('Al menos un elemento en la columna A es None')
      indices_nulos = change.index[change['target'].isnull()].tolist()
      print(indices_nulos)
      change.drop(indices_nulos, inplace=True)

  else:
      print('No hay elementos None en la columna A')
#df_mental_health_2 , df_fitness_2 , df_fitness_3
```

## Correcting hashtags

```python
import re
```

```python
list_of_dataframes = [df_mental_health , df_fitness , df_disorders_1 , df_disorders_2 , df_mental_health_2 , df_fitness_2 , df_fitness_3 , df_disorders_3]
for dataframe in list_of_dataframes:
  dataframe['hashtags'] = dataframe['hashtags'].apply(lambda x: re.findall(r'#\w+(?=_)', x)[0]) #podemos aplicar tranquilamente
```

```python
for change in list_of_dataframes:
  if change['hashtags'].isna().any():
      print('Al menos un elemento en la columna A es None')
  else:
      print('No hay elementos None en la columna A')
```
