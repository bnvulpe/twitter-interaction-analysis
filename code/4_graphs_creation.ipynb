## Combinaciones grafos

```python
list_of_dataframes = [df_mental_health , df_fitness , df_disorders_1 , df_disorders_2 , df_mental_health_2 , df_fitness_2 , df_fitness_3 , df_disorders_3]
```

```python
df_all_dataframes = pd.concat(list_of_dataframes, ignore_index=True)
```

```python
df_all_dataframes
```

```python
# agregar columna nueva
df_all_dataframes['initial_community'] = 0
disorders =['#ricecaketwt', '#proana', '#thinspo', '#anorexia','#eatingdisorder', '#anatwt','#edtwtdiet','#bulimia','#meanspo','#promia','#anamia','#EdTwitter','#edtwt','#bonespo','#eatingdisorders','#edtwtthread','#3dtwt']
health = ['#shtwt','#depression','#mentalhealth']
fitness = ['#healthy','#fitspo','#weightloss','#workout','#diet','#lowcal','#caloriedeficit','#fitnessmotivation']

# recorrer filas del DataFrame y agregar valores a la nueva columna según el valor de la columna "hashtags"
for i, row in df_all_dataframes.iterrows():
    if row['hashtags'] in disorders:
        df_all_dataframes.at[i, 'initial_community'] = 3
    if row['hashtags'] in health:
      df_all_dataframes.at[i, 'initial_community'] = 2
    if row['hashtags'] in fitness:
        df_all_dataframes.at[i, 'initial_community'] = 1
    #else:
      #df_all_dataframes.at[i, 'initial_community'] = 2
```

```python
df_all_dataframes['initial_community'].unique()
```

```python
df_all_dataframes_weighted = df_all_dataframes.groupby(["source", "target"]).weight.sum().reset_index()
df_all_dataframes_weighted
```

```python
df_all_dataframes_weighted_hash = df_all_dataframes.groupby(["source", "target", 'initial_community']).weight.sum().reset_index()
df_all_dataframes_weighted_hash
```

```python
from networkx.algorithms.assortativity.mixing import node_attribute_xy
all_communities_graph = nx.from_pandas_edgelist(df_all_dataframes_weighted, source="source", target="target", edge_attr="weight", create_using=nx.Graph())
```

```python
all_communities_graph_hash = nx.from_pandas_edgelist(df_all_dataframes_weighted_hash, source="source", target="target", edge_attr=["weight",'initial_community'], create_using=nx.Graph())
```

## Guardar grafos formato gephi

```python
from networkx.readwrite import gexf
gexf.write_gexf(all_communities_graph, "all_communities_graph.gexf")
```

```python
gexf.write_gexf(all_communities_graph_hash, "all_communities_graph_hash.gexf")
```
